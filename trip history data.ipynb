```python
import pandas as pd
import numpy as np
import datetime
```

```python
#2017 merge 
january = "resources/2017-trip-data/JC-201701-citibike-tripdata.csv"
february = "resources/2017-trip-data/JC-201702-citibike-tripdata.csv"
march = "resources/2017-trip-data/JC-201703-citibike-tripdata.csv"
april = "resources/2017-trip-data/JC-201704-citibike-tripdata.csv"
may = "resources/2017-trip-data/JC-201705-citibike-tripdata.csv"
june = "resources/2017-trip-data/JC-201706-citibike-tripdata.csv"
july = "resources/2017-trip-data/JC-201707-citibike-tripdata.csv"
august = "resources/2017-trip-data/JC-201708 citibike-tripdata.csv"
september = "resources/2017-trip-data/JC-201709-citibike-tripdata.csv"
october = "resources/2017-trip-data/JC-201710-citibike-tripdata.csv"
november = "resources/2017-trip-data/JC-201711-citibike-tripdata.csv"
december = "resources/2017-trip-data/JC-201712-citibike-tripdata.csv"

# put into dataframes
january = pd.read_csv(january)
february = pd.read_csv(february)
march = pd.read_csv(march)
april = pd.read_csv(april)
may = pd.read_csv(may)
june = pd.read_csv(june)
july = pd.read_csv(july)
august = pd.read_csv(august)
september = pd.read_csv(september)
october = pd.read_csv(october)
november = pd.read_csv(november)
december = pd.read_csv(december)

#merge seasons
winter_trip_history_2017 = pd.merge(january, february, how="outer")
winter_trip_history_2017 = pd.merge(winter_trip_history_2017, march, how="outer")

spring_trip_history_2017 = pd.merge(april, may, how="outer")
spring_trip_history_2017 = pd.merge(spring_trip_history_2017, june, how="outer")

summer_trip_history_2017 = pd.merge(july, august, how="outer")
summer_trip_history_2017 = pd.merge(summer_trip_history_2017, september, how="outer")

fall_trip_history_2017 = pd.merge(october, november, how="outer")
fall_trip_history_2017 = pd.merge(fall_trip_history_2017, december, how="outer")

#change object to datetime 
winter_trip_history_2017['Start Time']=pd.to_datetime(winter_trip_history_2017['Start Time'])
spring_trip_history_2017['starttime']=pd.to_datetime(spring_trip_history_2017['starttime'])
summer_trip_history_2017['starttime']=pd.to_datetime(summer_trip_history_2017['starttime'])
fall_trip_history_2017['starttime']=pd.to_datetime(fall_trip_history_2017['starttime'])
```

```python
#change winter column headers
winter_trip_history_2017 = winter_trip_history_2017.rename(columns={"Trip Duration":"tripduration", 
                                         "Start Time":"starttime", 
                                         "Stop Time":"stoptime", 
                                         "Start Station ID": "start station id",
                                         "Start Station Name": "start station name", 
                                         "Start Station Location":"start station location", 
                                         "Start Station Latitude":"start station latitude", 
                                         "Start Station Longitude": "start station longitude", 
                                         "Bike ID":"bikeid", 
                                         "User Type":"usertype", 
                                         "Birth Year":"birthyear", 
                                         "Gender":"gender"})
```

```python
#merge
citibike_2017_1 = pd.merge(winter_trip_history_2017, spring_trip_history_2017, how="outer")
citibike_2017_2 = pd.merge(summer_trip_history_2017, fall_trip_history_2017, how="outer")
citibike_2017 = pd.merge(citibike_2017_1, citibike_2017_2, how="outer")
```
